## City and country

```python
latitudes, longitudes = [], []

for i, row in subset[['city', 'country']].iterrows():
    location = str(row['city']) + ', ' + str(row['country'])
    geocoder = geopy.geocoders.Nominatim(user_agent='pdm', timeout=10)
    query_result = geocoder.geocode(location)
    # print(query_result)
    
    if query_result == None:
        latitudes.append(np.nan)
        longitudes.append(np.nan)
    else:
        latitudes.append(query_result.latitude)
        longitudes.append(query_result.longitude)
```
